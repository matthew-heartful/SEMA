# Initialization file

### Let's install some dependencies, the first line though is to restart the notebook, just in case you need to start over

```
import os
import subprocess

# Check if the requirements.txt file exists
if os.path.exists("./requirements.txt"):
    # Attempt to remove 'jedi' and 'jupyter' lines from requirements.txt
    # This requires reading the file, filtering lines, and writing it back
    with open("./requirements.txt", "r") as file:
        lines = file.readlines()
    with open("./requirements.txt", "w") as file:
        for line in lines:
            if 'jedi' not in line and 'jupyter' not in line:
                file.write(line)
                
    # Install the remaining requirements using pip
    subprocess.run(["pip", "install", "-r", "./requirements.txt"])
else:
    print("There's no requirements.txt, so nothing to install. This is the case with most projects.")
```
